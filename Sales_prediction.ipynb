Sales Prediction using Python

This project demonstrates the use of machine learning techniques for sales prediction. It includes data preprocessing, model training, prediction, and evaluation to understand how regression models can be applied to real-world business forecasting problems.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Advertising.csv")
dataset.head()
```

```python
dataset.columns=dataset.columns.str.strip()
```

```python
dataset.info
```

Dataset Understanding

The dataset contains advertising expenditure data for TV, Radio, and Newspaper platforms along with corresponding sales values. The Sales column is the target variable used for prediction.

```python
dataset.describe()
```

```python
dataset=dataset.drop(['Unnamed: 0'],axis=1)
```

Data Cleaning

The unnecessary index column was removed from the dataset because it does not contribute to the prediction process.

```python
x=dataset.drop('Sales',axis=1)
y=dataset['Sales']
```

Feature Selection

Advertising expenditure columns were selected as input features, while the Sales column was selected as the target variable for prediction.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

The dataset was divided into training and testing sets. The training set is used for learning patterns, while the testing set is used to evaluate model performance.

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
print("Model trained successfully")
```

Linear Regression Model

A Linear Regression model was trained to understand the relationship between advertising expenditure and sales performance.

```python
y_pred=model.predict(X_test)
print(y_pred[:5])
```

The trained model successfully predicted sales values for the test dataset based on advertising expenditure data.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print("MAE:",mean_absolute_error(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
```

Model Evaluation

The evaluation metrics indicate that the regression model achieved reasonable prediction accuracy with relatively low error values.

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted Sales")
plt.show()
```

The scatter plot shows a strong positive relationship between actual and predicted sales values. Most predicted values are close to the actual sales values, indicating effective model performance.

Conclusion

This project demonstrates how machine learning regression models can be used for sales prediction based on advertising expenditure data. Linear Regression was used to analyze the relationship between advertising investment and sales outcomes.
